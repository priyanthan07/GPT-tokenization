```python
def get_stats(ids):
    counts = {}
    for pair in zip(ids, ids[1:]):   # get consecutive pairs
        counts[pair] = counts.get(pair, 0) +1
    return counts

stats = get_stats(tokens)
print(sorted(((v,k) for k,v in stats.items()), reverse = True))
```

```python
top_pair = max(stats, key=stats.get)
top_pair
```

### merge tokens

```python
def merge(tokens, pair, new_index):
    new_encoding = []
    i = 0
    while i < len(tokens):
        if (i< len(tokens)-1) and (tokens[i] == pair[0]) and (tokens[i+1] == pair[1]):
            new_encoding.append(new_index)
            i += 2
        else:
            new_encoding.append(tokens[i])
            i += 1
    return new_encoding

new_tokens = merge(tokens, top_pair, 256)
    
print(new_tokens)

print("length :", len(new_tokens))
```

```python
vocab_size = 280
no_of_merges = vocab_size -256
org_encoding = list(tokens)

merges = {}
for i in range(no_of_merges):
    stats = get_stats(tokens)
    top_pair = max(stats, key=stats.get)
    index = 256 + i
    print(f"merge : {top_pair} -> {index}")
    tokens = merge(tokens, top_pair, index)
    merges[top_pair] = index
    
```

## Encoding

```python
def encode_tokens(text):
    tokens = list(text.encode('utf-8'))
    while len(tokens) >= 2:
        stats = get_stats(tokens)
        pair = min(stats, key = lambda x : merges.get(x, float("inf")))
        if pair not in merges:
            break
        index = merges[pair]
        tokens = merge(tokens, pair, index)
    return tokens
        
```

```python
print("encoding :", encode_tokens("f"))
```

## Decoding

```python
vocab = {idx: bytes([idx]) for idx in range(256)}
for (p0, p1), idx in merges.items():
    vocab[idx] = vocab[p0] + vocab[p1]
    
def decode_tokens(tokens):
    b_tokens = b"".join(vocab[idx] for idx in tokens)
    text = b_tokens.decode("utf-8", errors="replace")
    return text
```
